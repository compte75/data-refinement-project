```python
import pandas as pd
import numpy as np

# Chargement du brut
df = pd.read_csv('../DATA/RAW/dirty_cafe_sales.csv')

# --- Ã‰TAPE 1 : RÃ‰PARATION DES TYPES ---
print("--- Ã‰TAPE 1 : Conversion numÃ©rique ---")
numeric_cols = ['Quantity', 'Price Per Unit', 'Total Spent']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Tableau de rÃ©sultat 
res1 = pd.DataFrame({'Indicateur': ['Erreurs numÃ©riques traitÃ©es'], 'Nombre': [df['Total Spent'].isna().sum()]})
display(res1)

# RECALCUL 
print("\n--- Ã‰TAPE 2 : Recalcul du Chiffre d'Affaires ---")
price_map = df.groupby('Item')['Price Per Unit'].median().to_dict()
df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Item'].map(price_map))
df['Quantity'] = df['Quantity'].fillna(1)

# On recalcule le Total pour Ã©craser les anciennes "ERROR"
df['Total Spent'] = df['Quantity'] * df['Price Per Unit']

# Tableau de rÃ©sultat Ã‰tape 2
display(pd.DataFrame({'Action': ['Recalcul complet du Total Spent'], 'Statut': ['TerminÃ© âœ…']}))

# harmonisation des textes
print("\n--- Ã‰TAPE 3 : Nettoyage des textes ---")
for col in ['Item', 'Payment Method', 'Location']:
    df[col] = df[col].replace(['ERROR', 'UNKNOWN'], np.nan).fillna('Non SpÃ©cifiÃ©')

# DÃ‰DUPLICATION 
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')
df = df.dropna(subset=['Transaction Date']).drop_duplicates()

# ENREGISTREMENT DU FICHIER NETTOYÃ‰
df.to_csv('../DATA/PROCESSED/cafe_sales_cleaned.csv', index=False)
print("\nðŸ’¾ Fichier NETTOYÃ‰ enregistrÃ© dans DATA/PROCESSED/cafe_sales_cleaned.csv")
```
