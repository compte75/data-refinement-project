```python
import pandas as pd

# Chargement du fichier NETTOYÃ‰
df = pd.read_csv('../DATA/PROCESSED/cafe_sales_cleaned.csv')
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])

#ENRICHISSEMENT 
print("--- Ã‰TAPE 1 : Ajout de nouvelles colonnes ---")
df['Jour_Semaine'] = df['Transaction Date'].dt.day_name()
df['Mois'] = df['Transaction Date'].dt.month_name()
df['Type_Commande'] = df['Quantity'].apply(lambda x: 'Grosse' if x > 3 else 'Petite')

display(df[['Transaction ID', 'Item', 'Jour_Semaine', 'Type_Commande']].head())

#TABLEAUX DÃ‰CISIONNELS (Le rÃ©sultat pour ton rapport) 
print("\n--- Ã‰TAPE 2 : Performance par Produit ---")
product_table = df.groupby('Item')['Total Spent'].sum().sort_values(ascending=False).to_frame(name='CA Total')
display(product_table)

print("\n--- Ã‰TAPE 3 : Analyse de l'Affluence ---")
traffic_table = df.groupby('Jour_Semaine')['Transaction ID'].count().to_frame(name='Nombre de Clients')
display(traffic_table)

# ENREGISTREMENT DU FICHIER TRANSFORMÃ‰ (FINAL)
df.to_csv('../DATA/PROCESSED/cafe_sales_transformed.csv', index=False)
print("\nðŸ’¾ Fichier TRANSFORMÃ‰ enregistrÃ© dans DATA/PROCESSED/cafe_sales_transformed.csv")
```
