```python
import pandas as pd
import numpy as np

# Chargement
df_raw = pd.read_csv('../DATA/RAW/dirty_cafe_sales.csv')

print("--- TABLEAU D'AUDIT : VALEURS MANQUANTES ET ERREURS ---")
audit_data = {
    'Colonnes': df_raw.columns,
    'Valeurs Manquantes (NaN)': df_raw.isnull().sum().values,
    'Valeurs "ERROR"': [df_raw[col].astype(str).str.contains('ERROR').sum() for col in df_raw.columns],
    'Valeurs "UNKNOWN"': [df_raw[col].astype(str).str.contains('UNKNOWN').sum() for col in df_raw.columns]
}
print(pd.DataFrame(audit_data))
```
